```python
# imports

import numpy as np
import pandas as pd
from datetime import datetime, date, time, timedelta, timezone
from zoneinfo import ZoneInfo
```

```python
# create string that has date and time

now_string = str(datetime.now())   # or str(now.timestamp())
print(now_string, type(now_string))

bday_string = '1967-02-12 12:34:56'
print(bday_string, type(bday_string))
```

```python
# epoch start date and time l

epoch_plus_one = np.datetime64(1, 's')
print(epoch_plus_one, type(ep_plus_one))
```

```python
# pickup string and create datetime/timestamp/datetime64

now_datetime = datetime.strptime(now_string, '%Y-%m-%d %H:%M:%S.%f')
bday_datetime = datetime.strptime(bday_string, '%Y-%m-%d %H:%M:%S')
print(now_datetime, bday_datetime)

now_timestamp = pd.to_datetime(now_string)
bday_timestamp = pd.to_datetime(bday_string)
print(now_timestamp, bday_timestamp)

now_string_T = now_string.replace(' ', 'T')
bday_string_T = bday_string.replace(' ', 'T')

now_datetime64 = np.datetime64(now_string_T)      # must be ISO form YYYY-MM-DD or YYYY-MM-DDTHH:MM:SS
bday_datetime64 = np.datetime64(bday_string_T)     # must be ISO form YYYY-MM-DD or YYYY-MM-DDTHH:MM:SS
print(now_datetime64, bday_datetime64)
```

```python
# convert datetime/timestamp/datetime64 to string

print(str(now_datetime), str(now_timestamp), str(now_datetime64))

bday_datetime_string = datetime.strftime(bday_datetime, '%Y-%m-%d %H:%M:%S.%f')
bday_timestamp_string = bday_timestamp.strftime('%Y-%m-%d %H:%M:%S.%f')

bday_datetime64_string = bday_datetime64.astype(datetime)
bday_datetime64_string = bday_datetime64_string.strftime('%Y-%m-%d %H:%M:%S.%f')

print(bday_timestamp_string, bday_timestamp_string, bday_datetime64_string)
```

```python
# convert datetime/timestamp/datetime64 to date

now_datetime_date = datetime.date(now_datetime)
now_timestamp_date = now_timestamp.date()
now_datetime64_date = now_datetime64.astype('datetime64[D]')

print(now_datetime_date, now_timestamp_date, now_datetime64_date)

print(datetime.strftime(now_datetime_date, '%Y-%m-%d %H:%M:%S.%f'), 
      datetime.strftime(now_timestamp_date, '%Y-%m-%d %H:%M:%S.%f'))

now_datetime64_date = now_datetime64.astype('datetime64[D]')    # no good
now_datetime64_date = now_datetime64.astype('datetime64')
print(now_datetime64_date)

now_datetime64_date = now_datetime64.astype(datetime).date()    # good
print(datetime.strftime(now_datetime64_date, '%Y-%m-%d %H:%M:%S.%f'))
```

```python
# timezones

now_datetime_nyc = now_datetime.replace(tzinfo=ZoneInfo("US/Eastern"))
now_datetime_utc = now_datetime_nyc.astimezone(ZoneInfo("UTC"))

now_timestamp_nyc = now_timestamp.tz_localize("America/New_York")
now_timestamp_utc = now_timestamp_nyc.tz_convert("UTC")

#datetime64 doesn't really work with timezones - convert to datetime or timestamp

print(now_datetime_nyc, now_datetime_utc)
print(now_timestamp_nyc, now_timestamp_utc)
```
